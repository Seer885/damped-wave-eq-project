# Calculating derivatives numerically
# f(x) = x^3+x

```python
import numpy as np
import matplotlib.pyplot as plt

#User-defined function for f(x) and f'(x)
def f(x):
   return x**3+x

def ff(x):
    return 3*x**2+1

#Define our input values
a = 0
b = 10
n = 20
h = (b-a)/n
x_points = np.linspace(a, b, n+1)


#Define our initial output values and plot
y_points = f(x_points)
plt.plot(x_points, y_points, color="black")


#Define our derivative outputs and plot
yp_points = (f(x_points+h) - f(x_points)) / h

plt.plot(x_points, yp_points, color="red", linestyle="dotted")

#Plot the EXACT derivative for comparison purposes
d_points = ff(x_points)
plt.plot(x_points, d_points, color="red")


#Graph both of our functions
plt.grid()
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```
