```python
def f(x_val,expr):
    """
    Inputs:
        - Array of x-values
        - Function of x
    
    Outputs:
        -All corresponding y-values
    """
    import sympy as sp
    import numpy as np
    from sympy.abc import a,b,c,t,x,y,z # Import mathematical symbols for usage
    
    ## Enable python to recognize my function input as a computable function
    expr = sp.sympify(expr) # Convert the function into something computable
    
    func = sp.lambdify(x,expr, "numpy") # Allow the function to work with numpy concepts like arrays
    
    return func(x_val) # Return the output of replacing x with the x_val array

```

```python
import numpy as np
import matplotlib.pyplot as plt
x_val = np.linspace(0,10,10000)

y_val = f(x_val, "cos(x^2)-sin(x^3)")

# Graph our values
plt.figure(figsize=(10,7))
plt.plot(x_val, y_val)

plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import sympy as sp
import numpy as np
from sympy.abc import x

eqn = sp.Eq(x**3-2*x**2,-1)

solutions = sp.solve(eqn)
outputs = [round(sp.N(val),2) for val in solutions]

print(f'x = {outputs}')
```
